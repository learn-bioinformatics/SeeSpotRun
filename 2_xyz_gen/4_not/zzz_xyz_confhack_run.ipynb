```python
#Libraries
import sys
sys.path.append("../2_depend") # go to parent dir
from libsy import *
```

```python
xyz = pd.read_csv("../6_int/xyz_pos_obs.csv")
print("    in: "+"../6_int/xyz_pos_obs.csv")
xyz.head()
```

```python
multiplefiles = glob.glob("../3_input/confy/*dist.csv")
multiplefiles = sorted(multiplefiles)
multiplefiles
```

```python
conffy = []
for file in multiplefiles:
    dffy = pd.read_csv(file, sep=",", header=None)
    print("    in: "+file)
    ff = file
    ff = ff.replace("_conf_dist.csv","")
    ff = ff.replace("../3_input/confy/","")
    ff = ff.replace("_","")
    mdffy = pd.melt(dffy)
    mdffy['file'] = ff
    mdffy['deskc'] = mdffy['file'] + "_c" + mdffy['variable'].apply(str)
    # get index of timepoints
    mmm = []
    for m in mdffy['deskc'].unique():
        md = mdffy[mdffy['deskc'] == m]
        md = md.reset_index()
        md = md.drop(columns="index")
        md = md.reset_index()
        mmm.append(md)
    mmm = pd.concat(mmm)
    conffy.append(mmm)
conffy = pd.concat(conffy)
conffy['deskct'] = conffy['deskc'] + "_t" + conffy['index'].apply(str)
conffy = conffy.drop(columns=["index", "variable", "file", "deskc"])
```

```python
conffy.head()
```

```python
xyz_d = xyz
xyz_d['desk'] = xyz_d['locus'] + xyz_d['genotype'] + xyz_d['exp.rep'].apply(str) + xyz_d['meiosis']
xyz_d['deskc'] = xyz_d['desk'] + "_c" + (xyz_d['cell']-1).apply(str)
xyz_d['deskct'] = xyz_d['deskc'] + "_t" + (xyz_d['tp.n']-1).apply(str)
xyz_d.head()
```

```python
conffy.head()
```

```python
xyz_d.head()
```

```python
mrr = pd.merge(xyz_d, conffy, on=['deskct'], how='inner')
mrr = mrr.drop(columns=["desk", "deskc"])
mrr = mrr.rename(columns={"value": "confdist"})
mrr = mrr.sort_values(by=['locus', 'genotype', 'exp.rep', 'meiosis', 'cell', 'tp.n'])
mrr.head()
```

```python
# experiment removal
rem = pd.read_csv("../3_input/remove.csv")
print("    in: "+"../3_input/remove.csv")
mrr['desk'] = mrr['locus'] + "_" + mrr['genotype'] + "_" + mrr['exp.rep'].apply(str) + "_" + mrr['meiosis']
rmrr = mrr[-mrr['desk'].isin(rem['exp'])]
rmrr = rmrr.drop(columns=["deskct"])
rmrr.to_csv("../6_int/xyz_confhack_pre-exp.csv", sep='\t', index=False)
print("    out: "+"../6_int/xyz_confhack_pre-exp.csv")
rmrr.head()
```

```python
crmrr = rmrr
crmrr = crmrr.reset_index()
crmrr = crmrr.drop(columns=["index"])
crmrr = crmrr.reset_index()

start1 = time.time()

nac = crmrr
nac['X1'] = np.where(nac['confdist'].isnull(), float('nan'), nac['X1'])
nac['Y1'] = np.where(nac['confdist'].isnull(), float('nan'), nac['Y1'])
nac['Z1'] = np.where(nac['confdist'].isnull(), float('nan'), nac['Z1'])
nac['X2'] = np.where(nac['confdist'].isnull(), float('nan'), nac['X2'])
nac['Y2'] = np.where(nac['confdist'].isnull(), float('nan'), nac['Y2'])
nac['Z2'] = np.where(nac['confdist'].isnull(), float('nan'), nac['Z2'])

nac['X1'] = nac['X1']*0.1
nac['Y1'] = nac['Y1']*0.1
nac['Z1'] = nac['Z1']*0.1
nac['X2'] = nac['X2']*0.1
nac['Y2'] = nac['Y2']*0.1
nac['Z2'] = nac['Z2']*0.1

nxyz = nac

end1 = time.time()
print("time:")
print(end1 - start1)  
```

```python
nxyz.to_csv("../6_int/xyz_confhack_post-exp.csv", sep='\t', index=False)
print("    out: "+"../6_int/xyz_confhack_post-exp.csv")
nxyz.head()
```
