```python
#Libraries
import sys
sys.path.append("../2_depend") # go to parent dir
from libsy import *
```

```python
xyz = pd.read_csv("../6_int/comb_pos_xyz.csv")
print("    in: "+"../6_int/comb_pos_xyz.csv")
xyz['desc'] = xyz['locus'] + "_" + xyz['genotype'] + "_" + xyz['exp.rep'].apply(str) + "_" + xyz['meiosis']
xyz['desk'] = xyz['desc'] + "_c" + xyz['cell'].apply(str)
xyz.head()
```

```python
multifiles = glob.glob("../3_input/confy/*obs.xlsx")
multifiles = sorted(multifiles)
multifiles
```

```python
mcall = []
for file in multifiles:
    dffy = pd.read_excel(file, sep=",")
    print("    in: "+file)
    mcall.append(dffy)
mcall = pd.concat(mcall)
```

```python
obsy = mcall
obsy = obsy[obsy['Strain'].notnull()]
obsy = obsy[obsy['Strain'] != ""]
obsy.to_csv("../6_int/comb_obs.csv", sep=',', index=False)
print("    out: "+"../6_int/comb_obs.csv")
obsy.head()
```

```python
obs = pd.read_csv("../6_int/comb_obs.csv")
print("    in: "+"../6_int/comb_obs.csv")
obs['t'] = "t" + obs['t'].apply(str)
obs['desc'] = obs['Strain'] + "_" + obs['Cond'] + "_" + obs['Exp'].apply(str) + "_" + obs['t']
obs['desk'] = obs['desc'] + "_c" + obs['Cell'].apply(str)
obs = obs.sort_values(by=['Strain', 'Cond', 'Exp', 't', 'Cell'])
obs.head()
```

```python
ob = pd.DataFrame({"desk":obs['desk'], "Observations":obs['Observations']})
ob.head()
```

```python
xyz.head()
```

```python
xyzobs = pd.merge(xyz, ob, on=['desk'], how='inner')
xyzobs = xyzobs.rename(columns={"Observations": "observation"})
xyzobs.to_csv("../6_int/xyz_pos_obs.csv", sep=',', index=False)
print("    out: "+"../6_int/xyz_pos_obs.csv")
xyzobs.head()
```
